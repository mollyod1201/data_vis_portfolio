```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pregnancy data.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['AA_NOC'] = df['AA_NOC'].str.replace(',', '').astype(int)
```

```python
df['AA_PA'] = df['AA_PA'].astype(int)
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns

x1 = df['YOC']
y1 = df['AA_NOC']

#create scatterplot with regression line
p =sns.regplot(x1, y1, ci=None)

p.set_xlabel("Year of Conception", fontsize = 12)
p.set_ylabel("Number of Conceptions", fontsize = 12)
p.set_title("Number of Total Conceptions in Women Aged between 16 and 44, from 1990 to 2020", fontsize = 13)
```

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')
sns.relplot(data=df, x="YOC", y="AA_NOB", kind="line")
plt.title('Number of Total Births in Women Aged between 16 and 44, from 1990 to 2020')
plt.xlabel('Year of Conception')
plt.ylabel('Number of Births')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')
sns.relplot(data=df, x="YOC", y="AA_PA", kind="line")
plt.title('Proportion of Pregnancies resulting in Abortion in Women Aged between 16 and 44, from 1990 to 2020')
plt.xlabel('Year of Conception')
plt.ylabel('Proportion of Abortions')
plt.show()
```

```python
year3 = [1990, 2005, 2020]
years_3 = df2[df2['YOC'].isin(year3) ]
e = sns.catplot(data=years_3, x='YOC', y='PA', hue='Age', kind='bar', ci=None)
e.fig.suptitle('Proportion of Pregnancies resulting in Abortion in Women Aged between 16 and 44, in 1990, 2005 and 2020', fontsize=12);
plt.xlabel('Year of Conception')
plt.ylabel('Proportion of Abortions')
```
